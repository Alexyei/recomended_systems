```python
def convert_csv_to_df(df):
    df.drop(df.columns[[0]], axis=1, inplace=True)
    df.columns = np.arange(len(df.columns))
    Res = pd.DataFrame([],columns=['userId','movieId','rating'])
    for rowIndex, row in df.iterrows(): 
        for columnIndex, value in row.items():
#             print(rowIndex,columnIndex,value)
            if not np.isnan(value):
                Res = Res.append({'userId':columnIndex,'movieId': rowIndex, 'rating': value},ignore_index=True)
    return Res.astype(int);
```

```python
data1 = convert_csv_to_df(df)
data1.head()
```

```python
print(data1['userId'].unique().shape[0])
print(data1['movieId'].unique().shape[0])
```

```python
user_movie_rating1 = data1.pivot_table(index='userId', columns='movieId', values='rating')
user_movie_rating1
```

```python
user_movie_rating1.fillna(0.0, inplace=True)
user_movie_rating1
```

```python
## Вычисление предиктов
global_mean = user_movie_rating1[user_movie_rating1>0].mean().mean()
print(global_mean)
user_movie_rating_minus_mean = user_movie_rating1[user_movie_rating1>0]-global_mean
user_bias = user_movie_rating_minus_mean.T.mean()
print(user_bias)
item_bias = user_movie_rating_minus_mean.apply(lambda r: r - user_bias).mean()
print(item_bias)
```

```python
# заменяем нули предиктами
for rowIndex, row in user_movie_rating1.iterrows(): 
    for columnIndex, value in row.items():
        if (value == 0):
            user_movie_rating1.loc[rowIndex][columnIndex]=global_mean +user_bias[rowIndex]+item_bias[columnIndex]
print(user_movie_rating1)
```

```python
from numpy import linalg
U, Sigma, Vt = linalg.svd(user_movie_rating1)
```

```python
# print(Sigma)
print(Sigma.sum())
# 90%
print(Sigma.sum()*0.9)
print(Sigma[:6].sum())
```

```python
k=6
```

```python
def rank_k2(k):
    U_reduced= np.mat(U[:,:k])
    Vt_reduced = np.mat(Vt[:k,:])
    Sigma_reduced = Sigma_reduced = np.eye(k)*Sigma[:k]
    Sigma_sqrt = np.sqrt(Sigma_reduced)
    return U_reduced*Sigma_sqrt, Sigma_sqrt*Vt_reduced,Sigma_reduced
U_reduced, Vt_reduced,Sigma_reduced = rank_k2(4)
predicted_user_movie_rating1 = U_reduced * Vt_reduced
print(predicted_user_movie_rating1)
```

```python
# оценка пользователя 9 для фильма Аватар
print(predicted_user_movie_rating1[1,8])

print(predicted_user_movie_rating1[2,1])
print(predicted_user_movie_rating1[5,3])
print(predicted_user_movie_rating1[7,10])
print(predicted_user_movie_rating1[9,5])
```

```python
new_user = np.zeros(predicted_user_movie_rating1.shape[1])
print(new_user.shape)
# случайные оценки
new_user[0] = 5.0
new_user[3] = 3.0
new_user[10] = 5.0
```

```python
# делаем вменение
from numpy.linalg import inv
new_user_average = new_user[new_user > 0.0].mean()
print(new_user_average)
new_user[new_user == 0] = np.NaN
np.nan_to_num(new_user,copy=False,nan=new_user_average)
print(new_user)

new_user = new_user *Vt_reduced.T* inv(Sigma_reduced)
print(new_user)
```

```python
# Нормализуем оценки вычтя из оценки среднюю оценку пользователя
count_col = merged_user_ratings.groupby(['userId'])['rating'].count()
std_col = merged_user_ratings.groupby(['userId'])['rating'].std()
mean_col = merged_user_ratings.groupby(['userId'])['rating'].mean()
max_col = merged_user_ratings.groupby(['userId'])['rating'].max()
min_col = merged_user_ratings.groupby(['userId'])['rating'].min()

def rowFunc(row):
    return 0.0 if (count_col.at[row['userId']]==1) or (std_col.at[row['userId']]==0)  else(row['rating']-mean_col.at[row['userId']])/(max_col.at[row['userId']]-min_col.at[row['userId']])

merged_user_ratings['mean'] = merged_user_ratings.apply(rowFunc,axis=1)

 
merged_user_ratings.shape
```

```python
# проверяем, что нет пропусков
merged_user_ratings['mean'].isnull().any()
```

```python
users_values = merged_user_ratings['userId'].sort_values().unique()
movies_values = merged_user_ratings['movieId'].sort_values().unique()

user_categories = dict(enumerate(users_values))
movie_categories = dict(enumerate(movies_values))

def rowFunc(row):
    return np.where(movies_values == row['movieId'])[0][0], np.where(users_values == row['userId'])[0][0]

merged_user_ratings[['movieId','userId']] = merged_user_ratings.apply(rowFunc,axis=1,result_type="expand")
```

```python
from scipy.sparse import coo_matrix
coo = coo_matrix((merged_user_ratings['mean'], (merged_user_ratings['userId'],merged_user_ratings['movieId'])))
```

```python
coo.shape
```

```python
# cosine
M=coo.toarray()
print(M.shape)

def calculate_cosine_distance_cor(M):
    Res = np.empty((M.shape[0],M.shape[0]))
    def my_cosine_distance(i,j):
        return np.sum(M[i,:]*M[j,:])/(np.sqrt(np.sum(np.square(M[i,:])))*np.sqrt(np.sum(np.square(M[j,:]))))
    for i in range(M.shape[0]):
        for j in range(i,M.shape[0]):
            calculated = my_cosine_distance(i,j);
            if (i==j):
                Res[i,j]=calculated
            else:
                Res[i,j]=Res[j,i]=calculated
    Res[np.isnan(Res)] = 0
    return sparse.csr_matrix(Res)

cor = calculate_cosine_distance_cor(M)
print(cor.toarray())
```

```python
# Otiai
M=coo.toarray()


def calculate_otiai_distance_cor(M):
    Res = np.empty((M.shape[0],M.shape[0]))
    def my_otiai_distance(i,j):
        i_mean = np.sum(M[i,:])/(M[i,:]!=0).sum()
 
        j_mean = np.sum(M[j,:])/(M[j,:]!=0).sum() 

        i_minus_i_mean =  np.where((M[i,:]!=0), M[i,:]-i_mean, 0)
     
        j_minus_j_mean =  np.where((M[j,:]!=0), M[j,:]-j_mean, 0)
  
        numerator = np.sum(i_minus_i_mean*j_minus_j_mean)

        denumerator = (np.sqrt(np.sum(np.square(i_minus_i_mean)))*np.sqrt(np.sum(np.square(j_minus_j_mean))))
        return numerator/denumerator
    for i in range(M.shape[0]):
        for j in range(i,M.shape[0]):
            calculated = my_otiai_distance(i,j);
            if (i==j):
                Res[i,j]=calculated
            else:
                Res[i,j]=Res[j,i]=calculated
    Res[np.isnan(Res)] = 0
    return sparse.csr_matrix(Res)

cor = calculate_otiai_distance_cor(M)
print(cor.toarray())
```

```python
M=coo.toarray()
print(M.shape)

def calculate_pearson_distance_cor(M):
    Res = np.empty((M.shape[0],M.shape[0]))
    def my_pearson_distance(i,j):
        a = np.sum(M[i,:])/(M[i,:]!=0).sum()
        b = np.sum(M[j,:])/(M[j,:]!=0).sum()
        aa = np.where((M[i,:]!=0) &(M[j,:]!=0), M[i,:]-a, 0)
        bb = np.where((M[i,:]!=0) &(M[j,:]!=0), M[j,:]-b, 0)
        c = np.sum(aa*bb)
        d = (np.sqrt(np.sum(np.square(aa)))*np.sqrt(np.sum(np.square(bb))))
   
        return c/d
    for i in range(M.shape[0]):
        for j in range(i,M.shape[0]):
            calculated = my_pearson_distance(i,j);
            if (i==j):
                Res[i,j]=calculated
            else:
                Res[i,j]=Res[j,i]=calculated
    Res[np.isnan(Res)] = 0
    return sparse.csr_matrix(Res)

cor = calculate_pearson_distance_cor(M)
print(cor.toarray())
```

## Задача 4 (вручную)

```python
df4 = pd.read_csv("data3.csv",sep=";")
data4 = convert_csv_to_df(df4)
# Добавляем нового пользователя
columns = ['userId', 'movieId', 'rating']
new_user_ratings = pd.DataFrame([
    [11,0,4.0],
    [11,5,5.0],
    [11,6,2.0],
],columns=columns)

merged_user_ratings = pd.concat([data4,new_user_ratings],ignore_index=True,)
merged_user_ratings.tail(7)
```

```python
movie_data = merged_user_ratings 
```

```python
movie_data.loc[movie_data['userId']==0]
```

```python
movie_data.loc[(movie_data['movieId'] == 3) & (movie_data['userId'] == 0),'rating']
```

```python
for t in pd.unique(movie_data['userId']):
    user_movies = pd.Series(pd.unique(movie_data.loc[movie_data['userId']==t, 'movieId']))
    not_user_movies = pd.Series(pd.unique(movie_data[~movie_data['movieId'].isin(user_movies)]['movieId']))
    users_ratings_prediction = pd.DataFrame([],columns=['userId','movieId','rating'])

    users_mi_dict = {}
    for _, mi in user_movies.iteritems():
        users_mi_dict[mi] = movie_data.loc[movie_data['movieId'] == mi, 'userId']

    # для всех неоценённых пользователем фильмов
    for _, mj in not_user_movies.iteritems():
        R_j=0
        dev_j_i_sum=0

        users_mj = movie_data.loc[movie_data['movieId'] == mj,'userId']
    #     для всех оцененных пользователем фильмов
        for _, mi in user_movies.iteritems():

            users_mi = users_mi_dict[mi]

    #         ищем пользователем которые оценили оба фильма
            intersect_users = np.intersect1d(users_mj,users_mi)

    #     если такие пользователи нашлись
            if (intersect_users.size > 0):
                r_k_j = movie_data.loc[(movie_data['movieId'] == mj) & (movie_data['userId'].isin(intersect_users)),'rating'].sum()
                r_k_i = movie_data.loc[(movie_data['movieId'] == mi) & (movie_data['userId'].isin(intersect_users)),'rating'].sum()
    #             считаем среднюю разницу между оценками этих фильмов
                dev_j_i = (r_k_j-r_k_i)/intersect_users.size
    #     и прибавляем оценку текущего пользователя
                dev_j_i_sum += dev_j_i +  movie_data.loc[(movie_data['movieId'] == mi) & (movie_data['userId'] == t),'rating'].item()
                R_j+=1

    #         получаем среднюю оценку для неоценённого фильма по оценённым фильмам пользователя
        P_j = dev_j_i_sum/R_j if R_j > 0 else 0
        users_ratings_prediction.loc[users_ratings_prediction.shape[0]] = [t, mj, P_j]

print(users_ratings_prediction)
```
